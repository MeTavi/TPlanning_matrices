```python
def Tanner_func(x, alpha, beta):
    return (x ** beta) * np.exp(alpha * x)
```

```python
def EstimateParameters(func, x, y, p0=[-1,5]):
    '''Returns the parameters that optimize func so that 
    func(x,params) ~ y '''
    try:
        params, covs = optimize.curve_fit(func, x, y, p0=p0)
    except:
        params = (0, 0) #if estimation fails
    alpha, beta = params
    return alpha, beta
```
